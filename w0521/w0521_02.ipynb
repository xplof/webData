```python
import pandas as pd

df=pd.read_excel('data/score.xlsx',index_col='지원번호')
df
df['학년']=[1,2,3,2,3,1,2,3]
df
```

```python
df.groupby('학교').get_group('가산고')
df.groupby('학교').size()['가산고']
df.groupby('학교')[['키','국어']].mean()
```

```python
df.groupby('학교')[['키','국어','영어']].mean()
df.groupby(['학교','이름'])[['키','국어','영어']].mean()
df.groupby('이름')[['키','국어','영어']].mean().sort_values(['키','국어'],ascending=False)
```

```python
df.groupby('학교').count()
df['학교'].value_counts()
df.groupby('학교')['학년'].count().loc['가산고']
df.groupby('학교')['학년'].value_counts()
```

```python
df.groupby('학교')[['이름','SW특기']].count()
df.groupby('학교').value_counts()
df['학교'].value_counts()
df.value_counts()
df.groupby('학교')['학년'].value_counts()
df.groupby('학교')['키'].agg(['mean','max','min','sum']) #agg기본적인 통계정보 알려줌
df['키'].agg(['mean'])
df.groupby('학년')['키'].agg(['mean'])
```

```python
df['성별']=['남','여','남','남','남','남','여','여']
df
```

```python
#합계 컬럼 추가해서
#학교-성별별로 합계컬럼의 평균 최소 최대 합계점수 구하기
df['합계']=df['국어']+df['수학']+df['영어']+df['과학']+df['사회']
df
df.groupby(['학교','성별'])['합계'].agg(['mean','min','max','sum'])
```
